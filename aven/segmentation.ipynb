```python
from skimage import io
from skimage.color import rgb2gray
from skimage.segmentation import slic, mark_boundaries
from PIL import Image

image = io.imread('mang.jpg')

grayscale = rgb2gray(image)
segments = slic(grayscale, n_segments=100, compactness=10.0, enforce_connectivity=True, sigma=1)
image_segmented = mark_boundaries(grayscale, segments)
Image.fromarray(image_segmented)

```

```python
import cv2
import numpy as np

image = cv2.imread('mang.jpg')
grayscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, thresholded = cv2.threshold(grayscale, 100, 255, cv2.THRESH_BINARY)
thresholded = cv2.dilate(thresholded, None, iterations=2)
thresholded = cv2.erode(thresholded, None, iterations=1)
contours, _ = cv2.findContours(thresholded.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
for contour in contours:
    cv2.drawContours(image, [contour], -1, (0, 0, 255), 3)
cv2.imwrite('apples_segmented.jpg', image)
```

```python
import torch
import torchvision.transforms as T
from PIL import Image
from IPython.display import display
from matplotlib import pyplot as plt
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True) 
```

```python
transform = T.Compose([
    T.Resize(640),
    T.ToTensor()
])
```

```python
image_path = 'mang.jpg'
image = Image.open(image_path)
image = transform(image)
image = image.unsqueeze(0) 
```

```python
len(image[0][0])
```

```python
results = model(image)
results.print()
results.show()
segmented_image = results.render()
display(segmented_image)
```
